```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("equal.csv")

fig, ax = plt.subplots(figsize=(10, 5))
x=np.arange(4)
width = 0.35
y0 = df["Food picked 0"]
y1 = df["Food picked 1"]
labels = ("10/10", "10/20", "10/30", "10/40")
rects1 = ax.bar(x - width/2, y0, width, label='Colonie 0')
rects2 = ax.bar(x + width/2, y1, width, label='Colonie1')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Nourriture rapportée')
ax.set_title('Influence des proportions de fourmis dans les colonies')
ax.set_xlabel("Proportions de fourmis entre la colonie 0 et la colonie 1")
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```

```python
df = pd.read_csv("warrior.csv")

fig, ax = plt.subplots(figsize=(10, 5))
x=np.arange(len(df))
width = 0.35
y0 = df["Food picked 0"]
y1 = df["Food picked 1"]
labels = list(map(lambda x: x[1:2], df["n_warriors"].tolist()))
rects1 = ax.bar(x - width/2, y0, width, label='Colonie 0')
rects2 = ax.bar(x + width/2, y1, width, label='Colonie 1')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Nourriture rapportée')
ax.set_title(r'Influence du nombre de guerrières sur la performance')
ax.set_xlabel(r"Nombre de guerrières dans la colonie 0")
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```
